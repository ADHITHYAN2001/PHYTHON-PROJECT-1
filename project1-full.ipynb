## 5. Find if there is any correlation between age and salary , represent it visually.

```python
import seaborn as sns
```

```python
df1 =df.drop(["Name","Team","Number","Position","Height","Weight","College"],axis="columns")
corr = df1.corr()
corr
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation between age and salary")
plt.show()
```

```python
plt.scatter(df["Salary"],df["Age"],color="green")
plt.title("Correlation between age and salary")
plt.xlabel("Salary")
plt.ylabel("Age")
plt.legend()
plt.show
```
